```python
cato = ["race/ethnicity","test preparation course","lunch"]
for i in range(0,len(cato)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=data[cato[i]])
    plt.tight_layout()
```

```python
for i in range(0,len(numerical_features)):
    plt.subplot(2,2,i+1)
    sns.kdeplot(x=data[numerical_features[i]])
    plt.tight_layout()
```

```python
# box plotfor finding outliers
for i in range(0,len(numerical_features)):
    plt.subplot(2,2,i+1)
    sns.boxplot(x=data[numerical_features[i]])
    plt.tight_layout()
```

```python
# the total marks of male and female
data.groupby("gender")["Total_Marks"].sum().plot.pie(autopct="%1.1f%%",explode=(0.1,0))
```
